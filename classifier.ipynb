```python
data = np.loadtxt('/content/gdrive/MyDrive/AI_Hust_VLC /Du_lieu_gia_toc/datachuan/local_global_all/unnormal_feature_n8.csv', delimiter = ',')
X_raw = data[:,:-1]

#normalized du lieu ve mien [0;1]

# tạo bộ scaler
scaler =  MinMaxScaler() 
#fit scaler vao X_raw và thực hiện scale
X = scaler.fit_transform(X_raw)

y = data[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=200)
```

```python
#svm
# khởi tạo SVM classifier
clf = svm.SVC(kernel='linear') 
# Train classifier với dữ liệu
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

hs_svm = metrics.accuracy_score(y_test, y_pred)
print (round(100* hs_svm,2))
```

```python
def classifer (X , y, class_method = 'SVM', k = 2, h = 100, c = 11, g = 0.9875, kn = 'linear'):
    hs = []
    kf = KFold(n_splits=4)
    kf.get_n_splits(X)
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        if class_method == 'SVM' :
            clf = svm.SVC(C = c, kernel = kn, gamma = g) 
            clf.fit(X_train, y_train)
            y_pred = clf.predict(X_test)

            hs_svm = metrics.accuracy_score(y_test, y_pred)
            hs.append(round(100* hs_svm,1))
            cnf_matrix = confusion_matrix(y_test, y_pred)
            


        if class_method == 'KNN' :
            knn = KNeighborsClassifier(n_neighbors = k)
            knn.fit(X_train,y_train)
            y_pred = knn.predict(X_test)
            
            hs_knn = metrics.accuracy_score(y_test, y_pred)
            hs.append(round(100* hs_knn,1))
            cnf_matrix = confusion_matrix(y_test, y_pred)

        
        if class_method == 'RF' :
            rf = RandomForestClassifier(n_estimators = h)
            rf.fit(X_train, y_train)
            y_pred = rf.predict(X_test)

            hs_rf = metrics.accuracy_score(y_test, y_pred)        
            hs.append(round(100* hs_rf,1))
            cnf_matrix = confusion_matrix(y_test, y_pred)
      
    return hs, cnf_matrix

m, c = classifer(X, y, class_method = 'SVM',c = 1,g = 1/576)
print("Do chinh xac qua 4 lan danh gia: ")
print(m)
print("Ma tran nham lan: ")
print(c)
```
