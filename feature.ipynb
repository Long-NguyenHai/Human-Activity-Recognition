# Mục mới

```python
# làm mịn dữ liệu
def smooth(m):
  m_smooth = np.zeros(len(m))
  m_smooth = m
  for i in range(2 , (len(m_smooth) -2)):
    m_smooth[i] = (m_smooth[i-2] + m_smooth[i-1] +m_smooth[i] + m_smooth[i+1] + m_smooth[i+2])/5
  return m_smooth

# load file csv
def loaddata (namefile) : 
    time, Gx, Gy, Gz, Ax, Ay ,Az = np.loadtxt(namefile ,delimiter=',',unpack=True)
    mpre = Gx*Gx + Gy*Gy + Gz*Gz
    npre = Ax*Ax + Ay*Ay + Az*Az
    m = np.sqrt(mpre) #meanG
    n = np.sqrt(npre) #meanA
    Gx = smooth(Gx)
    Gy = smooth(Gy)
    Gz = smooth(Gz)
    m = smooth(m)
    Ax = smooth(Ax)
    Ay = smooth(Ay)
    Az = smooth(Az)
    n = smooth(n)
    return Gx, Gy, Gz, m, Az, Ay, Az, n

# phân đoạn thành n frame 
def framedata (data,n):
    valn = len(data) // n
    framedata = [0] * n
    for i in range(0,n):
        framedata[i] = data[i*valn : (i+1)*valn]

    return framedata
```

```python
# trích xuất đặc trưng
def f(data):
    meandata = np.mean(data)
    maxdata = np.max(data)
    mindata = np.min(data)
    stddata = np.std(data)
    energydata = np.sum(np.power(data,2))
    vardata = np.var(data)
    skewdata = skew(data)
    kurtdata = kurtosis(data)

    return [meandata, maxdata, mindata, stddata, energydata, vardata, skewdata, kurtdata]

os.chdir('/content/gdrive/MyDrive/AI_Hust_VLC /Du_lieu_gia_toc/Raw_Data/20211216_JAN_G_H/01')

# đọc dữ liệu từng người qua từng lần thực hiện 
data1 = loaddata ('01.csv')
data2 = loaddata ('02.csv')
data3 = loaddata ('03.csv')
data4 = loaddata ('04.csv')
data5 = loaddata ('05.csv')
data6 = loaddata ('06.csv')
data7 = loaddata ('07.csv')
data8 = loaddata ('08.csv')
data9 = loaddata ('09.csv')
data10 = loaddata ('10.csv')
data11 = loaddata ('11.csv')
data12 = loaddata ('12.csv')

# vtfeature để lưu các đặc trưng global và local của 1 lần thực hiện 1 hành động với n frame
def vtfeature(data , n) :
    feature = []
    for i in range (0 , len(data)):
        feature.extend(f(data[i])) #global

    frame_data = [0]*len(data)
    for i in range(0 , len(data)):
        frame_data[i] = framedata (data[i],n)
        for j in range (0,n): 
            feature.extend (f(frame_data[i][j])) #local
            
    return feature

# mảng lưu các vtfeature của 1 lần thực hiện với n frame 
def features (n):
    vt = []
    vt.append(vtfeature(data1 , n))
    vt.append(vtfeature(data2 , n))
    vt.append(vtfeature(data3 , n))
    vt.append(vtfeature(data4 , n))
    vt.append(vtfeature(data5 , n))
    vt.append(vtfeature(data6 , n))
    vt.append(vtfeature(data7 , n))
    vt.append(vtfeature(data8 , n))
    vt.append(vtfeature(data9 , n))
    vt.append(vtfeature(data10 , n))
    vt.append(vtfeature(data11 , n))
    vt.append(vtfeature(data12 , n))
    return vt

vt_n2 = features (2) # 2 frame
vt_n4 = features (4) # 4 frame
vt_n8 = features (8) # 8 frame

def savefile (vt , dir):
    saveFile = open (dir,'a')
    for i in range (0,len(vt)):
        for j in range(0,len(vt[i])):
            saveFile.write(str(vt[i][j]) + ',')
  
        saveFile.write(str(i))
        saveFile.write('\n')

os.chdir('/content/gdrive/MyDrive/AI_Hust_VLC /Du_lieu_gia_toc/datachuan/local_global_all')
savefile (vt_n2,'unnormal_feature_n2.csv')
savefile (vt_n4,'unnormal_feature_n4.csv')
savefile (vt_n8,'unnormal_feature_n8.csv')
```
